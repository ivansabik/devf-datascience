```python
import pandas as pd

pd.options.display.width = None
pd.options.display.max_columns = None

INPUT_PATH = '~/git/devf-datascience-17/ProcesarDatos1/Materiales/Inp/'
OUTPUT_PATH = '~/git/devf-datascience-17/ProcesarDatos1/Materiales/Out/'

df = pd.read_csv(INPUT_PATH + 'Import_venezuela.txt', sep='\t')

df.head()
```

## Quitar puntos, convertir a float

```python
del df['Unnamed: 12']
df = df.rename(columns={'Exportadores': 'country'})
for year in range(2001, 2012):
    col_name = 'Valor importada en {}'.format(year)
    df[col_name] = df[col_name].astype('string')
    df[col_name] = df[col_name].str.replace('.', '')
    df[col_name] = df[col_name].astype('float64')
    df = df.rename(columns={col_name: year})
    
df.head()
```

## Convertir de formato wide a long

```python
df = pd.melt(
    df,
    id_vars='country',
    var_name='year',
    value_name='imports',
    value_vars=range(2001, 2012)
)
df.head()
```

```python
df['imports-1'] = df['imports'].shift(1)

df.head()
```

```python
df.pivot_table(
    index=['country', 'year'],
    values=['imports'],
    aggfunc='mean'
)
```
