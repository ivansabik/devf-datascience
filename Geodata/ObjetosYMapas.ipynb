```python
fp = '/Users/britecorio/git/devf-datascience-17/Geodata/Mapa1/DesbymunTot.csv'
desparecidos = pd.read_csv(fp, dtype={'edo': 'object', 'mun': 'object'})
desparecidos['edo'] = desparecidos['edo'].str.zfill(2)
desparecidos['mun'] = desparecidos['mun'].str.zfill(3)
desparecidos['inegi'] = desparecidos['edo'] + desparecidos['mun']
desparecidos.head()
```

```python
desparecidos = desparecidos[['inegi', 'year', 'tdes', 'tdesh', 'tdesm']]
desparecidos.head()
```

```python
desparecidos_reshaped = desparecidos.pivot_table(
    index='inegi', columns='year', values=['tdes', 'tdesh', 'tdesm']
).reset_index()
desparecidos_reshaped.columns =[s1 + str(s2) for (s1,s2) in desparecidos_reshaped.columns.tolist()]
desparecidos_reshaped.head()
```

```python
desaparecidos_municipios = municipios[['inegi', 'geometry']].merge(
    desparecidos_reshaped,
    on='inegi',
    how='right'
)
desaparecidos_municipios.head(100)
```
